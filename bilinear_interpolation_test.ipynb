```python
def bilinear_interpolation(image, new_width, new_height):
    """
    Perform bilinear interpolation to resize the image to the specified resolution.

    :param image: 2D list representing the original image
    :param new_width: Desired width of the upscaled image
    :param new_height: Desired height of the upscaled image
    :return: 2D list representing the upscaled image
    """
    original_height = len(image)
    original_width = len(image[0])
    
    # Create an empty list for the new image
    upscaled_image = [[0 for _ in range(new_width)] for _ in range(new_height)]
    
    # Calculate scaling factors
    x_scale = (original_width - 1) / (new_width - 1)
    y_scale = (original_height - 1) / (new_height - 1)
    
    for i in range(new_height):
        for j in range(new_width):
            # Map the new pixel to the original image
            x = j * x_scale
            y = i * y_scale
            
            # Get the coordinates of the surrounding pixels
            x1 = int(x)
            y1 = int(y)
            x2 = min(x1 + 1, original_width - 1)
            y2 = min(y1 + 1, original_height - 1)
            
            # Calculate the weights for interpolation
            a = x - x1
            b = y - y1
            
            # Perform bilinear interpolation
            upscaled_image[i][j] = (
                (1 - a) * (1 - b) * image[y1][x1] +
                a * (1 - b) * image[y1][x2] +
                (1 - a) * b * image[y2][x1] +
                a * b * image[y2][x2]
            )
    
    return upscaled_image

# Example usage
new_width = 160  # Adjust as needed
new_height = 120  # Adjust as needed
upscaled_image = bilinear_interpolation(thermal_image, new_width, new_height)

# Visualize the upscaled image
plt.imshow(upscaled_image, cmap='hot', interpolation='nearest')
plt.colorbar(label='Temperature (°C)')
plt.title(f'Thermal Image Upscaled to {new_width}x{new_height}')
plt.xlabel('Width')
plt.ylabel('Height')
plt.show()
```
